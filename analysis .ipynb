## Using elbow method to find the required number of clusters

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


# Step 1: Text Vectorization
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['Cleaned_mails'].dropna())

# Step 2: Apply KMeans for a range of cluster numbers and calculate SSD
ssd = []
K = range(1, 11)  # Example range from 1 to 10
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    ssd.append(kmeans.inertia_)

# Step 3: Plot the SSDs
plt.figure(figsize=(10, 6))
plt.plot(K, ssd, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of Squared Distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
## Using Calinski Harabaz to evaluate the clusters by K mean for different number of clusters from 4,5,6,7
```

```python
from sklearn.metrics import calinski_harabasz_score

# Initialize an empty list to store Calinski-Harabasz Scores
ch_scores = []

# List of the number of clusters you want to test
cluster_numbers = [4, 5, 6, 7]

# Calculate Calinski-Harabasz Index for each number of clusters
for k in cluster_numbers:
    kmeans = KMeans(n_clusters=k, random_state=0).fit(X)
    labels = kmeans.labels_
    ch_score = calinski_harabasz_score(X.toarray(), labels)
    ch_scores.append(ch_score)
    print(f'Calinski-Harabasz Score with {k} clusters: {ch_score}')
```

Split your dataset into several subsets (folds).
Perform KMeans clustering on each subset.
Assess the stability of the clustering by comparing the cluster assignments across folds.

```python
from sklearn.model_selection import KFold
from sklearn.metrics import adjusted_rand_score
from sklearn.cluster import KMeans

# Initialize KFold
n_splits = 5
kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

# Number of clusters
n_clusters = 5

# This dictionary will hold the test indices and corresponding labels
fold_labels = {}

# Perform the KFold cross-validation
for fold, (train_index, test_index) in enumerate(kf.split(X)):
    # Split the data into training and test sets
    X_train, X_test = X[train_index], X[test_index]
    
    # Perform KMeans clustering on the training data
    kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)
    kmeans.fit(X_train)
    
    # Predict the clustering for the test set and store the labels with corresponding test indices
    fold_labels[fold] = (test_index, kmeans.predict(X_test))

# Calculate the stability of the clustering across folds using ARI
ari_scores = []
for i in range(len(fold_labels)):
    for j in range(i + 1, len(fold_labels)):
        # Get the intersection of indices between two folds
        common_indices = set(fold_labels[i][0]).intersection(set(fold_labels[j][0]))
        
        # Map these indices to positions in their respective label arrays
        common_pos_i = [pos for pos, index in enumerate(fold_labels[i][0]) if index in common_indices]
        common_pos_j = [pos for pos, index in enumerate(fold_labels[j][0]) if index in common_indices]
        
        # Retrieve the labels for these common positions
        labels_i = fold_labels[i][1][common_pos_i]
        labels_j = fold_labels[j][1][common_pos_j]
        
        # Compute ARI for these labels
        ari = adjusted_rand_score(labels_i, labels_j)
        ari_scores.append(ari)
        print(f"ARI between fold {i} and fold {j}: {ari}")

# Calculate the average ARI score across all fold comparisons
average_ari_score = sum(ari_scores) / len(ari_scores)
print(f"Average ARI score across all folds: {average_ari_score}")
```

BERT Tokenizer and Model Loading: The BERT tokenizer is used to convert email text into tokens that can be processed by the BERT model. BERT models are pre-trained on a large corpus of text and can be fine-tuned for a wide variety of NLP tasks. The bert-base-uncased model is a smaller BERT model that has been pre-trained on English language data.

Tokenization Function: The tokenize_email function takes an email's text as input and uses the BERT tokenizer to convert it into a format that the BERT model can understand (inputs). This typically includes token IDs, attention masks, and other necessary inputs for the model.

Tokenization of Email Content: Each email in the Cleaned_Body column of the dataframe df is tokenized using the tokenize_email function and the resulting data is stored in a new column called Tokenized_Email.

BERT Embeddings Extraction Function: The extract_bert_embeddings function processes the tokenized input through the BERT model (model(**email)) to obtain the embeddings. The last_hidden_state output is a sequence of hidden states representing each token in the input. The mean of these hidden states is taken (across the sequence dimension) to get a single vector representation for the entire email.

Extraction of BERT Embeddings for Each Email: The dataframe df is then updated with another new column called BERT_Embeddings, where each email's BERT embedding is stored. This is done by applying the extract_bert_embeddings function to the Tokenized_Email column.

K-Means Clustering: The num_clusters variable is set to the number of predefined categories you have. K-Means is initialized with this number of clusters and a random state for reproducibility. It is then fit to the list of BERT embeddings (email_embeddings) extracted from the emails.

Assignment of Cluster Labels: Finally, the resulting cluster labels from the K-Means clustering are added to the dataframe df in a new column called Cluster_Label

```python
# Define the categories
categories = ['Cyber Alerts', 'Service Alerts', 'Tender Opportunities', 'Event Notifications', 'Partner Newsletters', 'Social Media Alerts', 'Funding Opportunities', 'Online Magazines']

# Load the BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")

# Tokenize and prepare the data
def tokenize_email(email):
    inputs = tokenizer(email, return_tensors="pt", truncation=True, padding=True)
    return inputs

# Tokenize the email content and store it as a new column
df['Tokenized_Email'] = df['Cleaned_Body'].apply(tokenize_email)

# Extract BERT embeddings for each email
def extract_bert_embeddings(email):
    with torch.no_grad():
        outputs = model(**email)
    embeddings = outputs.last_hidden_state.mean(dim=1).squeeze().tolist()
    return embeddings

df['BERT_Embeddings'] = df['Tokenized_Email'].apply(extract_bert_embeddings)

# Perform K-Means clustering to classify emails
num_clusters = len(categories)  # Number of clusters based on predefined categories
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
email_embeddings = list(df['BERT_Embeddings'])
df['Cluster_Label'] = kmeans.fit_predict(email_embeddings)
```

```python
# Create a dictionary to hold dataframes for each cluster
clustered_dfs = {}

# Loop over each cluster number and create a separate dataframe
for cluster_num in range(num_clusters):
    clustered_dfs[cluster_num] = df[df['Cluster_Label'] == cluster_num]
```

using TF-IDF (Term Frequency-Inverse Document Frequency) to convert text data into a numerical format that reflects the importance of words in the emails. This numerical representation is then used to perform various analyses, including dimensionality reduction (with PCA) and identifying top words in the documents based on their TF-IDF scores.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import pandas as pd

def top_tfidf_feats(row, features, top_n=20):
    topn_ids = np.argsort(row)[::-1][:top_n]
    top_feats = [(features[i], row[i]) for i in topn_ids]
    return pd.DataFrame(top_feats, columns=['features', 'score'])

def top_feats_in_doc(X, features, row_id, top_n=25):
    row = np.squeeze(X[row_id].toarray())
    return top_tfidf_feats(row, features, top_n)
```

```python
# Iterate through each cluster
for cluster_num, cluster_df in clustered_dfs.items():
    # Vectorize the text data in the cluster
    vect = TfidfVectorizer(stop_words='english', max_df=0.50, min_df=2)
    X = vect.fit_transform(cluster_df['Cleaned_Body'])

    # Get feature names
    features = vect.get_feature_names_out()

    # For demonstration, let's analyze the first document in each cluster
    top_features_df = top_feats_in_doc(X, features, 0, 30)
    
    print(f"Cluster {cluster_num} Top Features:")
    print(top_features_df)
    print("\n")
```

```python
# Defining a function to extract from each entities
def extract_entities(text):
    doc = nlp(text)
    return [(ent.text, ent.label_) for ent in doc.ents]

```

##Applying NER to each cluster, computationally expensie process, After applying NER, each DataFrame will have a new column 'Entities' containing the named entities. We can now analyze these entities according to our requirements, such as counting the most frequent entities, examining types of entities, etc.

```python
for cluster_num, cluster_df in clustered_dfs.items():
    # Check if the column exists
    if 'Cleaned_Body' in cluster_df.columns:
        # Apply NER to each row in the DataFrame
        cluster_df['Entities'] = cluster_df['Cleaned_Body'].apply(extract_entities)
    else:
        print(f"Column 'Cleaned_Body' not found in cluster {cluster_num}")
```

## Counting most frequent entities

```python
from collections import Counter

for cluster_num, cluster_df in clustered_dfs.items():
    # Flatten the list of entities for the entire cluster
    all_entities = [entity for sublist in cluster_df['Entities'].tolist() for entity in sublist]
    
    # Count the frequencies of each entity
    entity_freq = Counter(all_entities)

    # Display the most common entities
    print(f"Cluster {cluster_num} - Most Common Entities:")
    print(entity_freq.most_common(10))  # Adjust the number as needed
    print("\n")
```

## Now we are examinig the different types of entities

```python
for cluster_num, cluster_df in clustered_dfs.items():
    # Extract entity types
    entity_types = [ent_type for _, ent_type in (entity for sublist in cluster_df['Entities'].tolist() for entity in sublist)]
    
    # Count the frequencies of each entity type
    entity_type_freq = Counter(entity_types)

    # Display the most common entity types
    print(f"Cluster {cluster_num} - Most Common Entity Types:")
    print(entity_type_freq.most_common())  # You can limit the number as needed
    print("\n")
```

## To identify the unique identities, entities

```python
# Collect all entities in a set for each cluster
all_entities_by_cluster = {cluster_num: set(entity for sublist in cluster_df['Entities'].tolist() for entity in sublist) 
                           for cluster_num, cluster_df in clustered_dfs.items()}

for cluster_num, entities in all_entities_by_cluster.items():
    unique_entities = entities.copy()
    for other_cluster, other_entities in all_entities_by_cluster.items():
        if cluster_num != other_cluster:
            unique_entities -= other_entities
    
    print(f"Cluster {cluster_num} - Unique Entities:")
    print(unique_entities)
    print("\n")
```

## To take a data with 1000 emails from all the clusters mainaining the proportion of the emails in the dataset, 

```python
def calculate_emails_to_extract(total_emails_cluster):
    # Total number of emails across all clusters
    total_emails = sum(total_emails_cluster)

    # Proportion of each cluster and calculating the number of emails to extract
    emails_to_extract = [round((cluster_emails / total_emails) * 1000) for cluster_emails in total_emails_cluster]

    # Adjust for rounding errors
    while sum(emails_to_extract) != 1000:
        # Find the index of the cluster with the maximum or minimum extraction to adjust
        if sum(emails_to_extract) > 1000:
            index = emails_to_extract.index(max(emails_to_extract))
            emails_to_extract[index] -= 1
        else:
            index = emails_to_extract.index(min(emails_to_extract))
            emails_to_extract[index] += 1

    return emails_to_extract


total_emails_cluster = [352,400,156,115,305,298,158,490] 
emails_to_extract = calculate_emails_to_extract(total_emails_cluster)
emails_to_extract
```

```python
import re
from transformers import T5Tokenizer

def preprocess_email_for_t5(email_body, max_length=512):
    # Initialize the T5 tokenizer
    tokenizer = T5Tokenizer.from_pretrained('t5-base')

    # Remove common greeting texts
    greetings_pattern = r"Hi\s\w+|Hello\s\w+|Hope\syou\sare\sdoing\swell|How\sare\syou\sdoing|How\sis\sit\sgoing"
    email_body = re.sub(greetings_pattern, "", email_body, flags=re.IGNORECASE)

    # Standardize dates and times
    date_pattern = r"\b\d{1,2}[-/]\d{1,2}[-/]\d{2,4}\b"  # Matches dates like 01/01/2020, 1-1-20 etc.
    time_pattern = r"\b\d{1,2}:\d{2}\b"  # Matches times like 10:30
    email_body = re.sub(date_pattern, "<date>", email_body)
    email_body = re.sub(time_pattern, "<time>", email_body)

    # Remove email signatures and disclaimers
    email_body = re.sub(r"--\s*[\s\S]*$", "", email_body)

    # Standardize email addresses and URLs
    email_body = re.sub(r"\S+@\S+\.\S+", "<email>", email_body)
    email_body = re.sub(r"http\S+", "<url>", email_body)

    # Add task-specific prefix
    email_body = "summarize: " + email_body

    # Tokenize and truncate if necessary
    tokens = tokenizer.encode(email_body, add_special_tokens=True, truncation=True, max_length=max_length)

    # Decode tokens back to string
    preprocessed_email = tokenizer.decode(tokens, skip_special_tokens=True)

    return preprocessed_email

# Example usage
cleaned_body = "Hi abc our meeting is on 01/01/2020 at 10:30. Please be prepared..."
preprocessed_email = preprocess_email_for_t5(cleaned_body)
print(preprocessed_email)
```
